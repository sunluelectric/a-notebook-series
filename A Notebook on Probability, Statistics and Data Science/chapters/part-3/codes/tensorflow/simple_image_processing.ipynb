```python

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
print("TensorFlow version: {}".format(tf.__version__))
print("Is GPU enabled: {}".format(tf.test.is_gpu_available()))
```

```python
def read_and_decode(filename, dims):
    img = tf.io.read_file(filename)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.image.resize(img, dims)
    return img
```

```python
img = read_and_decode('C:\\Users\\sunlu\\Downloads\\resources\\cats_img\\00000001_000.jpg', [256, 256])
```

```python
img_numpy = img.eval(session=tf.compat.v1.Session())
plt.imshow(img_numpy)
```
